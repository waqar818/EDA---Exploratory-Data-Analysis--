## (C) Cleaning the Data 


#### 1. Dealing with missing value

##### By Dropping Column

```python
#Dropping column 
df.drop(columns='deck', inplace=True)
```

```python
df.sample(10)
```

```python
df.isnull().sum()
```

##### By Replacing the Mean with missing value (Works for numeric attribute)

```python
# Replacing missing value with mean of age
df['age'] = df.age.fillna(value = df['age'].mean())
```

```python
df.isnull().sum() / len(df) * 100
```

```python
df.sample(20)
```

#### By Replacing Mode with missing value ( Works for catagorical attributes)

```python
df['embarked'] = df.embarked.fillna(value = df['embarked'].mode()[0])
```

```python
df.isnull().sum() / len(df) * 100
```

```python
df['embark_town'] = df.embark_town.fillna(value = df['embark_town'].mode()[0])
```

```python
df.isnull().sum() / len(df) * 100
```

#### Correcting error in data


##### 1. Removing Outliers

#### BY Visualization and IQR method

```python
# For 'Age'
# Removing Outliers by visualization
import seaborn as sns

sns.boxplot(data=df, y='age', x='sex')
```

```python
# Removing outliers from age

# By IQR
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
IQR
#Calculating Lower and upper limits
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
#Removing outliers 
df = df[(df['age'] > lower) & (df['age'] < upper)]
```

```python
#Removing Outliers from fare by visualzation
sns.boxplot(data=df, x='who', y='fare')
```

```python
# Removing outliers from fare

# By IQR
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
IQR
#Calculating Lower and upper limits
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
df = df[(df['fare'] > lower) & (df['fare'] < upper)]
```

```python
sns.histplot(df['age'])
```

```python
sns.histplot(df['fare'])
```

##### By Z-score method


```python
# Remove Outliers from "age" by Z-SCORE method
from scipy import stats
z_score = np.abs(stats.zscore(df['age']))
threshold = 3
df = df[(z_score < threshold)]
```

```python
# Remove Outliers from "fare" by Z-SCORE method
from scipy import stats
z_score = np.abs(stats.zscore(df['fare']))
threshold = 3
df = df[(z_score < threshold)]
```

### Log Transformation


#### Note: We can actually do log transformation after dealing with missing values and before finding outliers, So There is a chance that outliers should remove automatically by log transformation. 


```python
boat = sns.load_dataset('titanic')
boat.head()
```

##### Normalizaing "fare" column using log transformation
##### Note: Log transformation is not suitable for age

```python
import numpy as np
boat['fare'] = boat['fare'].fillna(boat['fare'].mean())
boat['age'] = boat['age'].fillna(boat['age'].mean())

# Create an array with some data
boat['fare'] = np.log(boat['fare'])
# Log transformation is not suitable for age
# boat['age'] = np.log(boat['age'])



# Log-transform the data

boat.head()
```

```python
# Checking by Normalization 
sns.boxplot(boat, x='sex', y='fare')
```

##### Normalizaing "age" column using log Min - Max Scaling
##### Note: Log transformation is not suitable for age thats why we are applying Min - Max scaling on it 

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


# Select columns to normalize
col = ['age']

#Creating scaler function 
scaler = MinMaxScaler()

#Fit and transform the data on scaler
boat[col] = scaler.fit_transform(boat[col])

boat.head()
```

```python
# Checking normalization by visualization 
sns.histplot(boat['age'])
```

```python
#Analyzing the null values by visualizing ( Heatmap)
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
#Dropping the deck Column 
df.drop(columns="deck", inplace=True)
```

```python
#Visualizing again to check if deck is removed
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
sns.countplot(data=df, x="survived")
```

#### Data distribution

```python
# Data distribution
sns.histplot(data=df, x="age", bins=20)
```

```python
#Working on Box plot to analyze the outliers
sns.boxplot(data=df, x='survived', y='age', hue='who')
```

```python
#Scatter plot ( WE SHOULD TAKE NUMERIC VALUE AT BOTH SIDE)
sns.scatterplot(data=df, x='age', y='fare', hue='pclass')
#limit is applied to 300 so it shows fare to 300
plt.ylim(0,300)
```

```python
sns.pairplot(data=df, hue="survived")
```

```python
#Stacked Barchart/plot
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
#Swarm plot
sns.swarmplot(data=df, x='survived', y='age', hue="who")
```

```python
#Point plot
sns.catplot(x='pclass',y='fare', hue='embarked',kind='point',data=df)
```
